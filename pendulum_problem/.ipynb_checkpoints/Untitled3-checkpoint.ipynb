```python
class PIDControlAgent():
    def __init__(self, fs, kp=1.2, ki=1.0, kd=0.001, set_angle=0.0):

        super(PIDControlAgent, self).__init__()
        self.kp = kp
        self.ki = ki
        self.kd = kd

        self.set_angle = set_angle
        self.tau = 1.0/fs

        self.p_term = 0.0
        self.i_term = 0.0[ -34.3875    ,  -47.32375   , -174.27239583,  -41.46286458]
        self.d_term = 0.0

        # cachematrix([[ -1.        ,  -2.24571918, -31.21627317,  -7.11387543]])
        self.output = 0.0
        self.err_prev = 0.0

    def update(self, v_in, v_fb):
        err = v_in - v_fb

        self.p_term = err
        self.i_term += err*self.tau
        self.d_term = (err - self.err_prev)*self.tau
        self.output = self.kp*self.p_term + self.ki*self.i_term + self.kd*self.d_term
        self.err_prev = err
        return self.output
    
    def choose_action(self, val):
#         if val >=0:
#             action = 0
#         else:
#             action = 1
        action = val
        print(val)
        return action

    def solving_act(self, state):
        output = self.update(self.set_angle, state[2])
        temp = self.choose_action(output)
        self.action = temp
        return self.action
```

### Full State Feedback  (FSF)

```python
class FSFAgent():
    def __init__(self, K, target_position=np.array([0.0, 0.0, 0.0, 0.0]) ):

        super(FSFAgent, self).__init__()
        self.K = K
        self.target_position = target_position
        # cache
        self.output = 0.0[ -34.3875    ,  -47.32375   , -174.27239583,  -41.46286458]

    def update(self, current_state):
        self.output = np.dot(self.K, current_state)
        #print('here', self.output)
        return self.output 
    
    def choose_action(self, val):
        
        action = val
#         if val >= 1:
#             action = 1
#         elif val <= -1:
#             action = -1
                
        return action

    def solving_act(self, state):
        output = self.update(np.array(state))
        temp = self.choose_action(output)
        self.action = temp
        return self.action
```

```python
imgs = []
actions_s = []
for i in count():
    next_action = agent.solving_act(current_state)
    # make a step
    obs, rew, done, _ = env.step([-next_action])#env.step(np.random.randint(0,2))# env.step(next_action)
    current_state = obs
    actions_s.append(next_action)
    # render images
    im = env.render(mode='rgb_array')
    imgs.append(im)
    if done or i >= 400:
        break
```
