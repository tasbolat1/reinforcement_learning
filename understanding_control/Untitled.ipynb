```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft
import matplotlib
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 22}

matplotlib.rc('font', **font)
```

```python
f1 = 10 # frequency of the signal 1
f2 = 5 # frequency of the signal 2
f3 = 1 # frequency of the signal 3

time_length = 10 # length of the signal
N = 1000 # number of samples from the signal within time_length
f = N/time_length # sampling frequency
t = np.linspace(0, time_length, N) # 
y = np.sin(2*np.pi*f1*t)+np.sin(2*np.pi*f2*t)+np.sin(2*np.pi*f3*t)
```

```python
# let's define this function in frequency domain
yf = fft(y)
xf = np.linspace(0.0, f/2.0, N//2)
```

```python
fig, ax = plt.subplots(2, figsize=(8,8))
ax[0].plot(t,y)
ax[0].set_ylabel(r'f(t)')
ax[0].set_xlabel('time')
ax[1].plot(xf, 2.0/N * np.abs(yf[0:N//2]))
ax[1].set_ylabel(r'F(f)')
ax[1].set_xlabel('frequencies')
plt.tight_layout(True)
plt.savefig('ex1.eps', dpi=100)
plt.show()
```
